### Exercise 11

```python
results["ex11_grade12_income"] = reduce_df[reduce_df["educ"] == "grade 12"][
    "inctot"
].mean()
reduce_df[reduce_df["educ"] == "grade 12"]["inctot"].mean()
```

### Exercise 12

```python
results["ex12_college_income"] = reduce_df[reduce_df["educ"] == "4 years of college"][
    "inctot"
].mean()
reduce_df[reduce_df["educ"] == "4 years of college"]["inctot"].mean()
```

```python
results["ex12_college_income_pct"] = (
    results["ex12_college_income"] / results["ex11_grade12_income"]
) * 100
results["ex12_college_income"] / results["ex11_grade12_income"] * 100
```

Our conclusion is that someone employed, whose highest education level is 4 years in college, is earning 193.76% of what someone employed, whose highest education level is grade 12, is earning.
