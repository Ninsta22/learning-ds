```python
print(
    "First, we are going to create the crime statistics for 2018. After, we merge both datasets, and verify that we properly performed a 1:1 merge"
)
new_merge["violent_arrest_rate_2018"] = (
    new_merge["VIOLENT"] / new_merge["total_population"]
)
new_merge["drug_arrest_rate_2018"] = (
    new_merge["F_DRUGOFF"] / new_merge["total_population"]
)
new_half = new_merge[["COUNTY", "violent_arrest_rate_2018", "drug_arrest_rate_2018"]]

old_half = testing_merge[
    ["COUNTY", "violent_arrest_rate_2009", "drug_arrest_rate_2009"]
]
both_years = pd.merge(
    old_half, new_half, how="outer", on="COUNTY", validate="1:1", indicator=True
)
both_years._merge.value_counts()
```

### Exercise 19

### Exercise 20

```python
average_drug_2009 = both_years["drug_arrest_rate_2009"].mean()
high_drug = both_years[both_years["drug_arrest_rate_2009"] > average_drug_2009]
low_drug = both_years[both_years["drug_arrest_rate_2009"] < average_drug_2009]
```

### Exercise 21

```python
results["ex21_diffindiff"] = (
    (
        high_drug["violent_arrest_rate_2018"] - high_drug["violent_arrest_rate_2009"]
    ).mean()
) - (
    (low_drug["violent_arrest_rate_2018"] - low_drug["violent_arrest_rate_2009"]).mean()
)
print(f"The difference in difference is {results['ex21_diffindiff']}")
```

### Exercise 22

With a difference in difference being negative, it means that there is a smaller change in those who were affected with high drug rates in 2009 than compared to counties with lower drug arrests in 2009. This seems to imply that legalization didn't have as big of a chance as expected.

### Exercise 23

```python
high_mean = (
    (
        high_drug["violent_arrest_rate_2018"] - high_drug["violent_arrest_rate_2009"]
    ).mean()
) / high_drug["violent_arrest_rate_2009"].mean()
low_mean = (
    (low_drug["violent_arrest_rate_2018"] - low_drug["violent_arrest_rate_2009"]).mean()
) / low_drug["violent_arrest_rate_2009"].mean()

results["ex23_diffindiff_proportionate"] = (high_mean - low_mean) * 100
print(
    f"The average county experienced a {results['ex23_diffindiff_proportionate']}% change in the violent arrest rate"
)
```
